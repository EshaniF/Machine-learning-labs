```python
from sklearn.cluster import KMeans
clustersum = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(df)
    clustersum.append(kmeans.inertia_)
plt.plot(range(1, 11), clustersum)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('clustersum')
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 3, random_state = 42)
y_kmeans = kmeans.fit_predict(df)
```

```python
plt.scatter(df[y_kmeans == 0, 0], df[y_kmeans == 0, 1], s = 60, c = 'red', label = 'Cluster1')
plt.scatter(df[y_kmeans == 1, 0], df[y_kmeans == 1, 1], s = 60, c = 'blue', label = 'Cluster2')
plt.scatter(df[y_kmeans == 2, 0], df[y_kmeans == 2, 1], s = 60, c = 'green', label = 'Cluster3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 100, c = 'black', label = 'Centroids')
plt.legend() 
plt.show()
```

Hierarchical Clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('malldata.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
newData = df.iloc[:, [3,4]].values
#newData
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
newData = sc.fit_transform(newData)
#newData
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = dendrogram(linkage(newData, method='ward'))
plt.axhline(y=30, color='r', linestyle='--')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')  
y_hc=cluster.fit_predict(newData)
y_hc
```

```python
plt.scatter(newData[y_hc == 0, 0], newData[y_hc == 0, 1], s = 60, c = 'red', label = 'Cluster1')
plt.scatter(newData[y_hc == 1, 0], newData[y_hc == 1, 1], s = 60, c = 'blue', label = 'Cluster2')
plt.scatter(newData[y_hc == 2, 0], newData[y_hc == 2, 1], s = 60, c = 'green', label = 'Cluster3')
plt.scatter(newData[y_hc == 3, 0], newData[y_hc == 3, 1], s = 60, c = 'cyan', label = 'Cluster4')
plt.scatter(newData[y_hc == 4, 0], newData[y_hc == 4, 1], s = 60, c = 'black', label = 'Cluster5')
plt.legend() 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("malldata.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
x = df.iloc[:, [3,4]].values
```

```python
from sklearn.neighbors import NearestNeighbors 
nearneigh = NearestNeighbors(n_neighbors=2) 
neighbrs=nearneigh.fit(x) 
distances,indices=neighbrs.kneighbors(x)
```

```python
distances = np.sort(distances, axis = 0) 
distances = distances[:, 1] 
plt.rcParams['figure.figsize'] = (5,3) 
plt.plot(distances) 
plt.show()
```

```python
from sklearn.cluster import DBSCAN
# cluster the data into five clusters
dbscan = DBSCAN(eps = 7, min_samples = 4).fit(x) 
labels = dbscan.labels_ 
```

```python
plt.scatter(x[:, 0], x[:,1], c = labels, cmap= "plasma") 
plt.xlabel("Income") 
plt.ylabel("Spending Score") 
plt.show() 
```
